```python
#Standardizing single letters to more readable word format
app1.CODE_GENDER= app1.CODE_GENDER.apply(lambda x: 'Male' if x=='M' else 'Female')
app1.FLAG_OWN_CAR= app1.FLAG_OWN_CAR.apply(lambda x: 'Yes' if x=='Y' else 'No')
app1.FLAG_OWN_REALTY= app1.FLAG_OWN_REALTY.apply(lambda x: 'Yes' if x=='Y' else 'No')
app1.head()
```

```python
#Since 'Female' is cleary considered the mode in the column, we can replace the 'XNA' values with 'Female' without affecting our
#analysis. We can also do this since the no. of rows having XNA is only 4.
#Another approach can be to drop the rows completely or treat them as NaN but I have chosen to replace them with the mode
app1.CODE_GENDER.value_counts()
```

```python
#creating a new column for age in years which is derived from DAYS_BIRTH column and converting it into positive numbers
app1['YEARS_BIRTH']= app1.DAYS_BIRTH.apply(lambda x: round(x*(-1)/365))
```

```python
#CLeaning the other columns to get positive numbers
app1.DAYS_EMPLOYED=app1.DAYS_EMPLOYED.apply(lambda x: x*(-1) if x<0 else x)
app1.DAYS_ID_PUBLISH=app1.DAYS_ID_PUBLISH.apply(lambda x: x*(-1))
app1.DAYS_REGISTRATION=app1.DAYS_REGISTRATION.apply(lambda x: x*(-1))
app1[['YEARS_BIRTH','DAYS_EMPLOYED','DAYS_ID_PUBLISH','DAYS_REGISTRATION']].head(10)
```

```python
#Checking to see if cleaning has worked as intended
print(app1[app1.DAYS_EMPLOYED<0].value_counts())
print(app1[app1.DAYS_ID_PUBLISH<0].value_counts())
print(app1[app1.DAYS_REGISTRATION<0].value_counts())
print(app1[app1.YEARS_BIRTH<0].value_counts())
```
